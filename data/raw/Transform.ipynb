# COMPREENSÃO DOS DADOS

* artist: nested dict contendo diversos links sobre o artista, como a URI (identificador universal) do artista além da URL que leva o usuário diretamente para a página do Spotify. * preview_url: contém uma prévia em formato mp3 de 30 segundos da música, o tipo de dado é object e o python traduz objetos vazios como 'None' e não 'Nan'. Também pode ser deletado

* track_number: Número da música dentro do álbum

* type: O tipo da música, sempre 'track'

* uri: URI que identifica a música unicamente

* is_local: Se a faixa é ou não de um arquivo local.

* album_name: Nome do álbum

* restrictions: Se há alguma restrição que impeça o usuário de acessar a música, pela minha avaliação, todas as músicas NÃO estão restritas, essa coluna também será deletada
